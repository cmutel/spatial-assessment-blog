```python
import matplotlib as mpt
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
```

```python
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[])
sankey = Sankey(ax=ax, unit=None)
sankey.add(
           facecolor='#37c959',
           flows=[1, -1.],
           labels=[None, r'$Avail_{0}$'],
           orientations=[0, 0])
sankey.add(facecolor='#58b1fa',
           flows=[1., -0.2, -0.8],
           labels=[None, r'$Loss_{1}$', r'$Avail_{1}$'],
           orientations=[0, 1, 0], prior=0, 
           connect=(1,0)
)
sankey.add(facecolor='#ff5555',
           flows=[0.8, -0.2, -0.6],
           labels=[None, r'$Use_{2}$', r'$Avail_{2}$'],
           orientations=[0, 1, 0], prior=1, 
           connect=(2,0)
)
diagrams = sankey.finish()
for diagram in diagrams:
    diagram.text.set_fontsize('14')
    for text in diagram.texts:
        text.set_fontsize('14')
plt.axis('off')
plt.savefig("sankey-elec.png", dpi=300, bbox_inches='tight', pad_inches=0)
```
